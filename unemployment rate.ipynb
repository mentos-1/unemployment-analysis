```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('/content/Unemployment_Rate_upto_11_2020[1].csv')
```

```python
fig = plt.figure(figsize =(15, 8))
plt.barh(df['Region'], df[' Estimated Unemployment Rate (%)'])
plt.show()
```

```python
fig = plt.figure(figsize =(15, 8))
plt.scatter(df['Region'], df[' Estimated Labour Participation Rate (%)'])
plt.title("Estimated Labour Participation Rate")
plt.show()
```

```python
fig = plt.figure(figsize =(25, 8))
plt.scatter(df['Region'], df[' Estimated Unemployment Rate (%)'])
plt.title("Estimated Unemployment Rate")
plt.show()
```

```python
fig = plt.figure(figsize =(16, 8))
sns.boxplot(data=df, x=df[' Estimated Unemployment Rate (%)'], y=df['Region'])
plt.show()
```

```python
fg = px.bar(df,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',title='Unemployment rate',animation_frame=' Date',template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

```python
sns.pairplot(data=df, hue='Region')
plt.show()
```

```python
fig = plt.figure(figsize =(10, 10))
fig = px.pie(df, values=" Estimated Unemployment Rate (%)", names="Region", title='Unemployment Rate')
fig.show()
```
